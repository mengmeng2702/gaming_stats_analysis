```python
import json
import pandas as pd 
import re
```

```python
with open('games.json', encoding="utf8") as f:
    data = json.load(f)

print(data)
```

```python
compact_list = []
simple_list = []

for i in data[0].keys():
    if type(data[0][i])==list:

        compact_list.append(i)

    else:
        simple_list.append(i)

compact_list
```

```python
compact_list.remove('achievements')
compact_list.append('rating')

simple_list.remove('url')
simple_list.remove('rating')
```

```python
rating_list = list()
genre_list = list()
oses_list = list()
languages_list = list()
n_reviews = list()


for game in data:
    genre_list.extend(game['genres'])
    oses_list.extend(game['oses'])
    for l in game['languages']:
        languages_list.append(l["name"])
    n_reviews.append(len(game['reviews']))
    try:
        rating_list.append(game["rating"])
    except:
        continue

rating_list = list(set(rating_list))
genre_list = list(set(genre_list))
genre_list.remove('')
oses_list = list(set(oses_list))
languages_list = list(set(languages_list))

rating_list
```

```python
dict1 = dict()
for s in simple_list:
    dict1[s] = []
    
dict1
```

```python
a = 0
for game in data:
    a+=1
    print(f'row {a}')
    for s in simple_list:
        try:
            dict1[s].append(game[s])
        except:
            dict1[s].append(None)

# dict1
```

```python
len(dict1['size'])
```

```python
df = pd.DataFrame.from_dict(dict1)

df
```

```python
age_res = list()
res_type = str()

for rating in rating_list:
    m1 = re.search('\d\d*',rating)
    age_res.append(m1.group(0))

    m2 = re.search('(?<=\().*(?=\))',rating)
    try:
        res_type = res_type + m2.group(0) + ', '
    except:
        continue

res_type = set(res_type.split(", "))
res_type = list(res_type)
res_type.remove('')

age_res = list(set(age_res))
```

```python
dict2 = dict()
for i in age_res:
    dict2[i] = []

for i in res_type:
    dict2[i] = []

for i in genre_list:
    dict2[i] = []

for i in ['Mac','Windows','Linux']:
    dict2[i] = []

for i in languages_list:
    dict2[i] = []
```

```python
a = 0
for game in data:
    a += 1
    print(f"row {a}")
    for i in age_res:
        try:
            if i in game['rating']:
                dict2[i].append(1)
            else:
                dict2[i].append(0)
        except:
            dict2[i].append(0)
    for i in res_type:
        try:
            if i in game['rating']:
                dict2[i].append(1)
            else:
                dict2[i].append(0)
        except:
            dict2[i].append(0)
    for i in genre_list:
        if i in game['genres']:
            dict2[i].append(1)
        else:
            dict2[i].append(0)
    for i in ['Mac','Windows','Linux']:
        c=0
        for j in game['oses']:
            if i in j:
                c+=1
            else:
                c+=0
        dict2[i].append(c)
    for i in languages_list:
        c=0
        for j in game['languages']:
            if i in j['name']:
                c+=1
            else:
                c+=0
        dict2[i].append(c)
```

```python
df2 = pd.DataFrame.from_dict(dict2)
df2
```

```python
df_tabular = pd.concat([df,df2], axis=1)
df_tabular
```
