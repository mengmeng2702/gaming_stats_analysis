# Load for_ml_model.csv to dataframe

```python
#### Read in and preview Gamespot Reviews
df = pd.read_csv('for_ml_model.csv')
print(df.shape)
df.head()
```

# Create new Dataframe for hours watched and hours streamed columns

```python
df_new = df[["tw_Hours_watched","tw_Hours_Streamed"]]
df_new
```

```python
# set index to Games for df_new 
df_new.set_index(df["Game"],inplace=True)
df_new
```

```python
#cluster the df_new
#standardize the data with standardscaler()
scaled_one = StandardScaler().fit_transform(df_new)
```

```python
df_scaled_one = pd.DataFrame(
    data=scaled_one, columns=["tw_Hours_watched","tw_Hours_Streamed"],index=df_new.index
)
df_scaled_one
```

```python
#create elbow curve to find the best K value
inertia = []
k = list(range(1,11))
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(df_scaled_one)
    inertia.append(km.inertia_)

elbow_data = {"k":k,"inertia":inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k",y="inertia",xticks=k,title= "Elbow Curve")
```

```python
# Initialize the K-Means model.
model = KMeans(n_clusters=3,random_state=0)

# Fit the model
model.fit(df_scaled_one)

# Predict clusters
predictions = model.predict(df_scaled_one)
predictions
```

```python
#scale data to create a scatter plot with hours
scaler = MinMaxScaler().fit_transform(df_new)
```

```python
# Create a new DataFrame that has the scaled data with the df_new DataFrame index.
plot_df = pd.DataFrame(
    data = scaler, columns=["Total Hours Watched","Total Hours Streamed"], index=df_new.index
)

# Add the "Game Name" column from the df DataFrame to the new DataFrame.
plot_df["Game Name"] = df_new.index

# Add the "Class" column from the clustered_df DataFrame to the new DataFrame. 
plot_df["Class"] = model.labels_

plot_df.head(10)
```

```python
plot_df.hvplot.scatter(x="Total Hours Streamed",y="Total Hours Watched",by="Class",hover_cols="Game Name")
```

```python
#getting games in a class
class_0_games = plot_df.loc[plot_df["Class"] == 0].index.unique().to_list()
class_1_games = plot_df.loc[plot_df["Class"] == 1].index.unique().to_list()
class_2_games = plot_df.loc[plot_df["Class"] == 2].index.unique().to_list()
```
