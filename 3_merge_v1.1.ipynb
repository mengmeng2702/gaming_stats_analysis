```python
import pandas as pd 
```

```python
df_review = pd.read_csv('steam_per_month_review_merged.csv')

# from https://www.kaggle.com/nikdavis/steam-store-games
df_games = pd.read_csv('data\steam.csv')
df_requirement = pd.read_csv('data\steam_requirements_data.csv')
df_category = pd.read_csv('data\steamspy_tag_data.csv')
```

```python
# merging the reviews data and basic information data together
df = df_review.merge(df_games, on='appid')
df
```

```python
# selecting columns to be kept
df = df[['time', 'appid', 'name', 'review', 'voted_up', 'release_date', 'english', 'developer', 'publisher', 'achievements']]
```

```python
df.head()
```

```python
# merging the dataframe from previous step with the requirement dataset
df = df.merge(df_requirement, left_on='appid', right_on='steam_appid')
df
```

```python
# dropping the redundant column of appid
df.drop(['steam_appid'], axis=1, inplace=True)
df
```

```python
# merge the category data with the main dataframe
df = df.merge(df_category, on='appid')
df
```

```python
# save the data to a csv file
df.to_csv('steam_cleaned_1.csv')
```
