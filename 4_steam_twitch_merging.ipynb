```python
# there seems to be a problem with the data type of the time column after loading the saved file back into a dataframe, so I have to turn it back into datetime type again
df['time']=pd.to_datetime(df['time'])
```

```python
# generating year column for the steam dataset
df['year']=df['time'].dt.year
```

```python
# month column
df['month']=df['time'].dt.month
```

```python
# drop the redundant unnamed column
df.drop(df.columns[0], axis=1, inplace=True)
```

```python
# merge the twitch data and steam data together by each game and each month
df_steam_twitch = df.merge(df_twitch, left_on=['name', 'year', 'month'], right_on=['Game', 'Year', 'Month'])
```

```python
# drop the redundant columns
df_steam_twitch.drop(['Game', 'Year', 'Month'], axis=1, inplace=True)
```

```python
df_steam_twitch.head(100)
```

```python
df_steam_twitch.isna().sum()
```

```python
# before generating system support columns with the requirement columns, checking how the data is formatted
# turns out that null data is an empty list
# It is reasonable to think that if the requirement is empty, then that game is not on that platform, and otherwise it is supported somehow on that platform
df_steam_twitch['linux_requirements'][0]=='[]'
```

```python
# the data in hours streamed column is written like: '__ Hours', so the 'Hours' part needs to be removed
hours_streamed = df_steam_twitch['Hours_Streamed']
```

```python
hours_streamed = hours_streamed.apply(lambda x : x.split(' ')[0])
```

```python
df_steam_twitch['Hours_Streamed'] = hours_streamed
```

```python
# generating system support columns
windows = df_steam_twitch['pc_requirements']
mac = df_steam_twitch['mac_requirements']
linux = df_steam_twitch['linux_requirements']
```

```python
windows = windows.apply(lambda x : 0 if x =='[]' else 1)
mac = mac.apply(lambda x : 0 if x =='[]' else 1)
linux = linux.apply(lambda x : 0 if x =='[]' else 1)
```

```python
df_steam_twitch['windows'] = windows
df_steam_twitch['mac'] = mac
df_steam_twitch['linux'] = linux
```

```python
# no null values in the dataframe
df_steam_twitch['minimum'].isna().sum()
```

```python
# drop unnecessary columns
df_steam_twitch.drop(['pc_requirements','mac_requirements','linux_requirements','minimum'], axis=1, inplace=True)
```

```python
# turn the recommended column into boolean
# the recommended column indicates whether there is a recommended hardware setting above the minimum requirement
df_steam_twitch['recommended']=df_steam_twitch['recommended'].fillna(0)
```

```python
df_steam_twitch['recommended'] = df_steam_twitch['recommended'].apply(lambda x : 1 if x != 0 else 0)
```

```python
max(df_steam_twitch.isna().sum())
```

```python
# save the dataframe to a new file
df_steam_twitch.to_csv('steam_twitch.csv')
```
