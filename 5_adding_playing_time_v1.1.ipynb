```python
# loading in the steam playtime data from https://www.kaggle.com/michau96/popularity-of-games-on-steam
df_playtime = pd.read_csv('SteamCharts.csv', encoding = "ISO-8859-1")
```

```python
# creating a function to turn the months into numerical values
def month_to_num(month):
    month = month.replace(' ', '')

    if month == 'January':
        a = 1
    elif month == 'February':
        a = 2
    elif month == 'March':
        a = 3
    elif month == 'April':
        a = 4
    elif month == 'May':
        a = 5
    elif month == 'June':
        a = 6
    elif month == 'July':
        a = 7
    elif month == 'August':
        a = 8
    elif month == 'September':
        a = 9
    elif month == 'October':
        a = 10
    elif month == 'November':
        a = 11
    elif month == 'December':
        a = 12
    else:
        print(month)
        raise Exception('oops')
    return a

```

```python
# create a function to turn the percentages into numerical values
def per_to_num(per):
    a = per.replace('%','')
    a = float(a)/100
    return a
```

```python
# turning month from string to int
df_playtime['month']=df_playtime['month'].apply(lambda x : month_to_num(x))
```

```python
# turning percentage from string to float
df_playtime['avg_peak_perc'] = df_playtime['avg_peak_perc'].apply(lambda x : per_to_num(x))
```

```python
# merging the main dataframe with the playtime dataframe
df = df.merge(df_playtime,left_on=['name', 'year', 'month'], right_on=['gamename', 'year', 'month'])
```
